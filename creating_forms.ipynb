```python


def plt_slices_sim(simulation):
    fig,axs = plt.subplots(1,3,figsize = (15,5))
    ax = axs[0]
    im = ax.imshow(np.log(simulation.FTI_custom[226,:,:]), cmap = 'Greys')
    ax = axs[1]
    im = ax.imshow(np.log(simulation.FTI_custom[:,226,:]), cmap = 'Greys')
    ax = axs[2]
    im = ax.imshow(np.log(simulation.FTI_custom[:,:,226]), cmap = 'Greys')
    plt.show()    
```
